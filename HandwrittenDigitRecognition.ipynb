```python
import tensorflow as tf
from tensorflow.keras.layers import Input,Flatten,Dense,Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.regularizers import l2
Seed=42
tf.random.set_seed(Seed)
```

# Loading the dataset

```python
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.mnist.load_data()
```

# Shape

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

# Visualization of first 10 images of Train Dataset

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))

for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(x_train[i],cmap="grey")
    plt.axis("off")
    plt.title(y_train[i])
plt.tight_layout()
plt.show()
```

# Building the Architecture of both the model

```python
#model1 has 2 hidden layer each consist of 64 neurons and with relu activation function and also performs l2 regularization to prevent overfitting  in it
model1=Sequential()
model1.add(Input(shape=(28,28)))
model1.add(Flatten())
model1.add(Dense(
    64,
    activation="relu",
    kernel_regularizer=l2(0.0001)
))

model1.add(Dense(
    64,
    activation="relu",
    kernel_regularizer=l2(0.0001)
))

model1.add(Dense(
    10,
    activation="softmax"
))
```

```python
#model2 has 2 hidden layer each consist of 64 neurons and with sigmoid activation function and also performs l2 regularization to prevent overfitting  in it
model2=Sequential()
model2.add(Input(shape=(28,28)))
model2.add(Flatten())
model2.add(Dense(
    64,
    activation="sigmoid",
    kernel_regularizer=l2(0.0001)
))

model2.add(Dense(
    64,
    activation="sigmoid",
    kernel_regularizer=l2(0.0001)
))

model2.add(Dense(
    10,
    activation="softmax"
))
```

# Compilation of both the model

```python
model1.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model2.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

# Training of both the model

```python
Seed=42
tf.random.set_seed(Seed)
hist_model1=model1.fit(x_train,y_train,epochs=20,batch_size=32,verbose=1,validation_split=0.2)
```

```python
Seed=42
tf.random.set_seed(Seed)
hist_model2=model2.fit(x_train,y_train,epochs=18,batch_size=32,verbose=1,validation_split=0.2)
```

# Printing Training Accuracy

```python
print("Training Accuracy\n")
train_acc={
    "model1":round(hist_model1.history["accuracy"][-1],3),
    "model2":round(hist_model2.history["accuracy"][-1],3)
}
for name,accuracy in train_acc.items():
    print(f"{name}:{accuracy}")
```

# Printing Validation Accuracy

```python
print("Validation Accuracy\n")
validate_acc={
    "model1":round(hist_model1.history["val_accuracy"][-1],3),
    "model2":round(hist_model2.history["val_accuracy"][-1],3)
}
for name,accuracy in validate_acc.items():
    print(f"{name}:{accuracy}")
```

# Testing of both the model

```python
test_model1=model1.evaluate(x_test,y_test,verbose=1)
```

```python
test_model2=model2.evaluate(x_test,y_test,verbose=1)
```

# Calculating Generation Gap

```python
# Generation Gap can be defined as difference between training accuracy and validation accuracy. i.e. GP=TA-VA
# It can be used as a indicator to determine whether the overfitting is present or not in the model
# If the value of generation gap is small that means the model is learning pattern from the dataset instead of mugging up

#Calculating Generation Gap
models=[
    "model1","model2"
]
train=[
   round(hist_model1.history["accuracy"][-1],3),
   round(hist_model2.history["accuracy"][-1],3)
]
validate=[
    round(hist_model1.history["val_accuracy"][-1],3),
    round(hist_model2.history["val_accuracy"][-1],3)
]
generation_gap=[round(train[i]-validate[i],3) for i in range(len(models))]
print(generation_gap)
```

```python
parameter={
    "model1":model1.count_params(),
    "model2":model2.count_params()
}
for name,par in parameter.items():
    print(f"{name}:{par}")
```

```python
parameters=[
    model1.count_params(),
    model2.count_params()
]
print(parameters)
```

```python
import pandas as pd
print(type(pd))
```

# Comparision matrix

```python
print("Comparison Matrix\n")
comp_mat=pd.DataFrame({
    "Models":models,
    "Train Accuracy":train,
    "Validation Accuracy":validate,
    "Generation Gap":generation_gap,
    "Total Parameters":parameters
})
print(comp_mat.set_index("Models"))
```

# Visualizing the training accuracy,validation accuracy per epoch

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,6))
# Training Accuracy vs Epoch
plt.subplot(1,2,1)
colors=['red',"green"]
plt.plot(hist_model1.history["accuracy"],label='model1',color=colors[0])
plt.plot(hist_model2.history["accuracy"],label="model2",color=colors[1])
plt.title("Training Accuracy per Epoch- Both Models")
plt.xlabel("Epochs")
plt.ylabel("Training Accuracy")
plt.legend()

# Validation Graph per Epoch
plt.subplot(1,2,2)
plt.plot(hist_model1.history["val_accuracy"],label="model1",color=colors[0])
plt.plot(hist_model2.history["val_accuracy"],label="model2",color=colors[1])
plt.title("Validation Accuracy per Epoch- Both Models")
plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.legend()



plt.tight_layout()
plt.show()
```

# Visualizing the testing accuracy and generation gap per models

```python
test=[test_model1[1],test_model2[1]]
plt.figure(figsize=(10,6))
# Testing Accuracy vs models
plt.subplot(1,2,1)
colors=['red',"green"]
bars=plt.bar(models,test,color=colors)
plt.title("Testing Accuracy per Model")
plt.xlabel("Models")
plt.ylabel("Testing Accuracy")
plt.legend(bars,models)

# Generation Gap vs models
plt.subplot(1,2,2)
colors=['red',"green"]
bar=plt.bar(models,generation_gap,color=colors)
plt.title("Generation Gap per Model")
plt.xlabel("Models")
plt.ylabel("Generation Gap")
plt.legend(bar,models)

plt.tight_layout()
plt.show()
```

# Learning Curves for both model

```python
plt.figure(figsize=(16,6))

plt.subplot(1,2,1)
colors=['red',"green"]
plt.plot(hist_model1.history["accuracy"],label='Train',color=colors[0])
plt.plot(hist_model1.history["val_accuracy"],label="Validation",color=colors[1])
plt.title("Accuracy Learning Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()



plt.subplot(1,2,2)
colors=['red',"green"]
plt.plot(hist_model2.history["accuracy"],label='Train',color=colors[0])
plt.plot(hist_model2.history["val_accuracy"],label="Validation",color=colors[1])
plt.title("Accuracy Learning Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.tight_layout()
plt.show()
```

# From the all above Graph i can say that no model is overfitting because:
1. The generation Gap for both the model is low
2. The learning curve of training and validation for both models is staying close and improving per epoch.But due to high validation accuracy than training accuracy, the model2 is slightly moving towards underfitting.

#By analyzing,all the above graph and comparison matrix I can say that the best model is model1.
#Reason for selecting model1 as best model:
1. model1 has high validation accuracy(0.956) as compared to model2(0.924) which plays a important factor in decision making.
2. model1 has also as high training accuracy(0.979) as compared to model2(0.923).
3. model1 has low and healthy generation gap(0.023) as compared to model2(-0.001) which means model2 is slightly underfitting.i.e.it has high validation accuracy than a training accuracy

# Predicting the output using the best model

```python
y_pred=model1.predict(x_test)
print(y_pred)
```

```python
print(y_pred.shape)
```

```python
print(y_pred[0])
```

# model1.predict() will always return probability distribution of all labels for that particular datapoint

# Converting probability distribution into labels

```python
import numpy as np
y_pred_labels=[np.argmax(i) for i in y_pred]
print(y_pred_labels)
```

# y_test-> True Labels
# y_pred_labels-> Predicted labels

```python
from tensorflow.math import confusion_matrix
cm=confusion_matrix(y_test,y_pred_labels)
print(cm)
```

# Visualizing tne confusion matrix

```python
# The Diagonal elements of the matrix shows the correct predictions
# While the off diagonal shows the error or the incorrect predictions
import seaborn as sns
plt.figure(figsize=(16,6))
sns.heatmap(cm,annot=True,fmt="d",cmap="Blues")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```

# Some Predictions for best model

```python
print("Format: True label/Predicted label")
correct_indices = np.where(y_pred_labels == y_test)[0]
plt.figure(figsize=(7,7))
plt.suptitle(f"Some Correct predictions of best_model: model1")
for plot_index in range(40):
    plt.subplot(4,10, plot_index+1)
    plt.imshow(x_test[correct_indices[plot_index]],cmap="grey")
    plt.title(f"{y_test[correct_indices[plot_index]]}/{y_pred_labels[correct_indices[plot_index]]}",fontsize=8)
    plt.axis("off")
plt.tight_layout()
plt.show()
```
